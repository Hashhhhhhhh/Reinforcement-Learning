```python
import os
import gymnasium as gym # Changed from gym to gymnasium
from stable_baselines3 import PPO #(model)
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
#loading the environments
env=gym.make('CartPole-v1')
```

```python
#testing the environment (test it 5 times)
tests=5
for i in range(1,tests+1):
  print(i)
```

```python
env.reset() #initial values
```

```python
env.action_space.sample() #0 or 1
```

```python
env.observation_space
```

```python
env.step(1) #reward = 1.0 #true or false tells if the episode is done or not. ## For CartPole, action 1 means moving right
```

```python
for i in range (1,tests+1): #tests/episodes
  state=env.reset()
  score=0
  done=False
  while not done:
    env.render()
    action=env.action_space.sample() #selects a random action from the set of valid actions
    n_state,reward,done,info=env.step(action)
    score+=reward
    print('Test:{} Score:{}'.format(i,score))
env.close()
```

Observation Space (env.observation_space): Represents the information the agent receives, such as pixel data, joint angles, or sensory data.

Action Space (env.action_space): Represents all possible moves or decisions an agent can make, such as moving left/right, torque values, or menu selections.

**COMMON SPACE TYPES :**

Discrete: A fixed range of non-negative integers (e.g., Discrete(4) for 4 distinct actions: 0, 1, 2, 3).

Box: A n-dimensional continuous space, representing bounded, numerical data (e.g., coordinates, pixel values).
Dict / Tuple: Combinations of multiple space types

```python
# shimmy is no longer needed when using gymnasium directly

# Corrected: Create a new environment instance using gymnasium.make
env=DummyVecEnv([lambda: gym.make('CartPole-v1')])
model=PPO('MlpPolicy',env)
model.learn(total_timesteps=20000)
```

**Testing and Evaluation**

Two core metrics has to be paid attention to than the other metrics available for training a model.


ep_len_mean and ep_rew_mean which tells how long a particular episode would last before getting done and the avg reward the agent accumulated per episode respectively.


```python
evaluate_policy(model,env,n_eval_episodes=18,render=True)
```

```python
env.close()
```

```python
#testing model
for i in range (1,tests+1):
  obs=env.reset()
  score=0
  done=False
  while not done:
    env.render()
    action,_=model.predict(obs)
    obs,reward,done,info=env.step(action)
    score+=reward
  print(f"Test:{i} Score:{score.item()}")
```

Callbacks are used to either log out of data or save the model at some particular instant. Keras callbacks are functions that automatically run at specific points during model training to monitor, save, or adjust your model.

```python
#callbacks
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold, StopTrainingOnNoModelImprovement, StopTrainingOnMaxEpisodes
stop_callback= StopTrainingOnRewardThreshold(reward_threshold=200)
# Corrected: Pass stop_callback to EvalCallback's callback_on_new_best argument
eval_callback=EvalCallback(env,best_model_save_path='./logs/', eval_freq=1000, n_eval_episodes=10, callback_on_new_best=stop_callback)
#apply callback to training model
model.learn(total_timesteps=20000,callback=[eval_callback])
```

```python
#changing policies
from stable_baselines3 import DQN
model=DQN('MlpPolicy',env,verbose=1)
model.learn(total_timesteps=20000,callback=[eval_callback])
```

```python
DQN_path = os.path.join('Training', 'Saved Models', 'DQN_Model_Cartpole')
model.save(DQN_path)
```
